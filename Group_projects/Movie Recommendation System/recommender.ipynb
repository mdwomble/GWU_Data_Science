### Handling Categorical Variables

Pulling data out from dictionaries and list for the following variables:
- Genre
- Production Companies
- Production Countries
- Spoken Languages
- Cast
- Crew
- Keywords

```python
# List of columns to parse names from
columns=['genres', 'production_companies', 'production_countries', 'spoken_languages', 'cast', 'crew', 'keywords']

# Cycle through columns and pull names
import ast
for column in columns:
  # Remove row with empty values
  df_movies = df_movies.loc[df_movies[column]!='[]']
  # remove dictionary values from list
  df_movies[column] = df_movies[column].apply(lambda x: x[1:-1])
  dicts =  []
  for dict in df_movies[column]:
    # print(dict)
    my_dict = ast.literal_eval(dict)
    # print(my_dict)
    dicts.append(my_dict)
  movies=[]
  for x in range(len(dicts)):
      names=[]
      movie=dicts[x]
      for y in range(len(movie)):
          if str(type(movie))!="<class 'tuple'>":
              name=movie['name']
          if str(type(movie))=="<class 'tuple'>":
              name=movie[y]['name']
          names.append(name)
      unique_names=[]
      for x in names:
        if x not in unique_names:
          unique_names.append(x)
      movies.append(unique_names)

  df_movies[column]=movies
```

One-hot-encoding multi-label categorical variables

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb=MultiLabelBinarizer()

# One-hot encode genre
genre_array=mlb.fit_transform(df_movies['genres'])
df_genres=pd.DataFrame(genre_array, columns=mlb.classes_)

# One-hot encode production_companies
production_companies=mlb.fit_transform(df_movies['production_companies'])
df_companies=pd.DataFrame(production_companies, columns=mlb.classes_)

# One-hot encode production_countries
production_countries=mlb.fit_transform(df_movies['production_countries'])
df_countries=pd.DataFrame(production_countries, columns=mlb.classes_)

# One-hot encode spoken_languages
languages=mlb.fit_transform(df_movies['spoken_languages'])
df_languages=pd.DataFrame(languages, columns=mlb.classes_)

# One-hot encode cast
cast_array=mlb.fit_transform(df_movies['cast'])
df_cast=pd.DataFrame(cast_array, columns=mlb.classes_)

# One-hot encode crew
crew_array=mlb.fit_transform(df_movies['crew'])
df_crew=pd.DataFrame(crew_array, columns=mlb.classes_)

# One-hot encode keywords
keywords_array=mlb.fit_transform(df_movies['keywords'])
df_keywords=pd.DataFrame(keywords_array, columns=mlb.classes_)
```

```python
# Rename columns
df_genres=df_genres.add_prefix('Genre_')
df_companies=df_companies.add_prefix('Company_')
df_countries=df_countries.add_prefix('Country_')
df_languages=df_languages.add_prefix('Language_')
df_cast=df_cast.add_prefix('Cast_')
df_crew=df_crew.add_prefix('Crew_')
df_keywords=df_keywords.add_prefix('Keywords_')
```
